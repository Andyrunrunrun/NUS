## 1. Introduction to Jupyter Notebook

Here we will cover just the basics of Jupyter Notebook. For more details please click [here](https://www.dataquest.io/blog/jupyter-notebook-tutorial/)

### 1.1 Cells

Jupyter Notebooks are organized into cells. Cells can be in one of two modes; Code mode (default) and Markdown mode. In code mode anything you type in a cell will be treated as Python code, whereas in Markdown mode anything you type will be treated as Markdown.

Mardown is a very powerful way of creating documents, and is supported on major platforms like github, and of course Jupyter Notebook itself.  You will even be able to write complex mathematical expressions like $z_i = \sum_{i=0}^n \frac{x_i + 1} {y_i - 1}$ easily.

### 1.2 Getting to Edit Mode

By default you will be in Command Mode when you enter a Jupyter Notebook. To edit a cell, click (or double-click if the cell is being run) the cell to edit it. Alternatively if you are at the cell already, just press enter. You will now be able to type normally in the cell.

There are several commands available in Edit mode. Please see [here](http://maxmelnick.com/2016/04/19/python-beginner-tips-and-tricks.html)

### 1.3 Getting to Command Mode

To get to command mode, press the ESC key. You then have the following commands available:

A Insert a cell above
B Insert a cell below
M Change cell to Markdown
Y Change cell to Code
D D Delete current cell (Press D twice)
C Copy cell
P Paste cell
SHIFT-M Merge selected cells or merge with cell below.
SHIFT-ENTER Run cell and select next cell below
CTRL-ENTER Run cell

Please look [here](http://maxmelnick.com/2016/04/19/python-beginner-tips-and-tricks.html) for more commands.

**To change back to Edit Mode press Enter.**

### 1.4 Markdown Cells

Markdown is a very powerful syntax for formatting text. Some Markdowns include:

# First level headings
## Second level headings
### Third level headings
#### Fourth level headings

You need to indent in order to get lists:

    - Unnumbered lists
    1. Numbered lists

*italics*
**bold**

Inline mathematical formulae like $x^2 = y^2 + z^2$

Block formulae like: $$x^2 = y^2 + z^2$$

Hyperlinks like [this](https://www.comp.nus.edu.sg), and much much more.

Code fences like:

```
print("Hello world!")
```

Please see [here](https://www.markdownguide.org/cheat-sheet/) for a convenient Markdown cheat sheet.

You can also escape special characters like \$
### Code cells

Code cells have an In \[ \] that shows you the number of times the code has run, and the output is shown following the code cell. You can type any valid Python code in code boxes, including docstrings. The important thing to understand is that code effects carry forward from the topmost cells to the bottom most. For example:


# Here we set x to 5
x = 5

Now in the cell below we can do:

print(x) # Will show 5 since this was set earlier

The other important thing to know is that you can start running cells from anywhere. Note that if you started running the code cell directly above (the one with 'print(x)') without running the earlier code ('x = 5'), you will get an error since you've not executed the assignment statement yet.
